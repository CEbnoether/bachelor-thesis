```python
#Logistic Regression - Subjective Efficiency H1
import pandas as pd
import statsmodels.api as sm

file_path = r'C:\Users\corsi\Documents\Universität\Bachelorarbeit\Experiment\DataForPythonLR_Subjective.xlsx'
data = pd.read_excel(file_path)

# H1 (1) with Control (0)
data_subset = data[data['Group'].isin([0, 1])].copy()  # Filter for control and H1 only

data_subset['Is_Treatment'] = data_subset['Group'].apply(lambda x: 1 if x == 1 else 0)

y = data_subset['subjective efficiency']

X = data_subset[['Is_Treatment', 'age','gender', 'disposable income', 'risk aversion', 'health state','chronic disease', 'hospitalization', 'mathematics', 'financial 3', 'expected medical costs', 'confidence medical costs']]  # Add any other control variables as needed
X = sm.add_constant(X)  # Adds an intercept to the model

logit_model = sm.Logit(y, X)
result = logit_model.fit()

print(result.summary())
```
