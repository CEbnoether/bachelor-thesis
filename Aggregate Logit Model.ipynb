```python
import pandas as pd
import statsmodels.api as sm

# Load the data
file_path = r'C:\Users\corsi\Documents\Universität\Bachelorarbeit\Experiment\DataAggregateLogisticModel.xlsx'
data = pd.read_excel(file_path)

# Define the dependent variable
y = data['subjective efficiency']

# Define independent variables (treatment dummies + controls)
X = data[['Dummy TH1', 'Dummy TH2', 'Dummy TH3', 
          'age', 'gender', 'disposable income', 'risk aversion', 
          'health state', 'chronic disease', 'hospitalization', 
          'mathematics', 'financial 3', 'expected medical costs', 
          'confidence medical costs']]

X = sm.add_constant(X)  # Add intercept

# Fit logistic regression model
logit_model = sm.Logit(y, X)
result = logit_model.fit()

# Show summary of results
print(result.summary())
```
